```python
df = df.drop(columns = ["Difference","Combined_Key","Admin2","iso2","iso3","Prep_Flow_Runtime","Data_Source"])
df
```

```python
df.isna().sum()
```

```python
df.dropna(subset = ["Province_State"],inplace = True)
df["People_Total_Tested_Count"].fillna(df["People_Total_Tested_Count"].mean(), inplace = True)
df["FIPS"].fillna(df["FIPS"].mean(), inplace = True)
df["Lat"].fillna(df["Lat"].mean(), inplace = True)
df["Long"].fillna(df["Long"].mean(), inplace = True)
df["Population_Count"].fillna(df["Population_Count"].mean(), inplace = True)
df["People_Hospitalized_Cumulative_Count"].fillna(df["People_Hospitalized_Cumulative_Count"].mean(), inplace = True)
df.isna().sum()
```

```python
df.Country_Region.unique()
```

```python
df.groupby("Date").describe()
```

```python
print("The average number of cases in various country")
df.groupby("Country_Region").agg({"Cases" : ['mean']})
```

```python
print("The number of case type based on population in various countries")
casetyp = (df.groupby(["Country_Region","Case_Type"])).agg({"Population_Count":['mean']})
casetyp
```

```python
print("The amount of samples tested on respective date")
Tested = df.groupby(["Country_Region","Date"]).agg({"People_Total_Tested_Count":['mean']})
Tested
```

```python
plt.figure(figsize=(20,6))
plt.title ('Visualization of Country_Region and cases')
sns.barplot(x=df['Country_Region'],y=df['Cases'])
```

```python
plt.figure(figsize=(10,6))
plt.title ('Visualization of casetypes as per population count')
sns.barplot(x=df['Case_Type'],y=df['Population_Count'])
```

```python
plt.figure(figsize = (10,6))
print("The count of people hozpitalized in different countries")
sns.barplot(x=df['Country_Region'],y=df['People_Total_Tested_Count'])
```

```python
plt.figure(figsize = (16,6))
#add a title
plt.title("Country vs People hospitalized")
print("The number of people Hospitalized out of tested counts")
sns.lineplot(x = df['People_Total_Tested_Count'],y=df['People_Hospitalized_Cumulative_Count'])
```
